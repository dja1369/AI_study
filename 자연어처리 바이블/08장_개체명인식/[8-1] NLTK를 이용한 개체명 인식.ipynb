```python
# -*- coding:utf-8 -*-
import nltk

nltk.download('punkt')
nltk.download('words')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
```

```python
sentence = "Prime Minister Margaret Thatcher has said in the past that Britain will be able to survive by 5 May."
```

```python
tokens = nltk.word_tokenize(sentence)
print(tokens)
```

```python
tagged = nltk.pos_tag(tokens)
print(tagged)
```

```python
entities = nltk.chunk.ne_chunk(tagged)
print(entities)
```
