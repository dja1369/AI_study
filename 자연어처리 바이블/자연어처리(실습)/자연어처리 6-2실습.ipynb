```python
# 영어 multi-task 통계 모델
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp('The fat cat sat on the mat')
```

```python
for token in doc:
  print(token.text, token.dep_, token.head.text)
```

```python
from spacy import displacy
```

```python
# Jupyter, Colab 등에서 동작
displacy.render(doc, style='dep', jupyter=True)
```
