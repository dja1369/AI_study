```python
# konlpy 관련 패키지 import
from konlpy.tag import Okt
from konlpy.tag import Kkma
from konlpy.tag import Hannanum
from konlpy.tag import Komoran
from konlpy.tag import Twitter
```

```python
kkma = Kkma()
okt = Okt()
komoran = Komoran()
hannanum = Hannanum()
twitter = Twitter()
```

```python
# konlpy 의 라이브러리 품사태깅 비교
print("okt 형태소 :", okt.pos(u"집에 가면 감자 좀 쪄줄래?"))
print("kkma 형태소 : ", kkma.pos(u"집에 가면 감자 좀 쪄줄래?"))
print("hannanum 형태소 : ", hannanum.pos(u"집에 가면 감자 좀 쪄줄래?"))
print("komoran 형태소 : ", komoran.pos(u"집에 가면 감자 좀 쪄줄래?"))
print("twitter 형태소 : ", twitter.pos(u"집에 가면 감자 좀 쪄줄래?"))
```
