```python
# konlpy 패키지 다운로드
!pip install konlpy

# konlpy 관련 패키지 import
from konlpy.tag import Okt
from konlpy.tag import Kkma
from konlpy.tag import Hannanum
from konlpy.tag import Komoran
from konlpy.tag import Twitter

kk = Kkma()
ok = Okt()
kom = Komoran()
han = Hannanum()
twit = Twitter()

# konlpy 의 라이브러리 품사태깅 비교
print("ok 형태소 :", ok.pos(u"피시방 에 언제 갈껀지 말해 줄래?"))
print("kk 형태소 : ", kk.pos(u"피시방 에 언제 갈껀지 말해 줄래?"))
print("han 형태소 : ", han.pos(u"피시방 에 언제 갈껀지 말해 줄래?"))
print("kom 형태소 : ", kom.pos(u"피시방 에 언제 갈껀지 말해 줄래?"))
print("twit 형태소 : ", twit.pos(u"피시방 에 언제 갈껀지 말해 줄래?"))
```
