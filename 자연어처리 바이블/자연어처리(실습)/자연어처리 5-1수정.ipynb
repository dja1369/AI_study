```python
!pip install konlpy

# konlpy 관련 패키지 import
from konlpy.tag import Okt
from konlpy.tag import Kkma
from konlpy.tag import Hannanum
from konlpy.tag import Komoran
from konlpy.tag import Twitter

kk = Kkma()
ok = Okt()
kom = Komoran()
han = Hannanum()
twit = Twitter()

#konlpy 중 Kkma는 문장 분리가 가능 (다른 라이브러리는 되지 않음)
print ("kkma 문장 분리 : ", kk.sentences('저는 운동을 좋아합니다 선생님도 좋아하세요?.'))

# konlpy 의 라이브러리 형태소 분석 비교
print("ok 형태소 분석 :", ok.morphs(u"집에 가면 프로틴 좀 타줄래?"))
print("kk 형태소 분석 : ", kk.morphs(u"집에 가면 프로틴 좀 타줄래?"))
print("han 형태소 분석 : ", han.morphs(u"집에 가면 프로틴 좀 타줄래?"))
print("kom 형태소 분석 : ", kom.morphs(u"집에 가면 프로틴 좀 타줄래?"))
print("twit 형태소 분석 : ", twit.morphs(u"집에 가면 프로틴 좀 타줄래?"))
```
