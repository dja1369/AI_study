```python
!pip install nltk==3.3

import nltk

word = nltk.CFG.fromstring("""
S -> NP VP
NP -> NP JX | MAG NNG JKS 
VP -> NP VP | VA EPT EFN

NP -> '나'
JX -> '는'
MAG -> '항상'
NNG -> '배'
JKS -> '가'
VA -> '고프'
EPT -> '았'
EFN -> '어요'
""")

parser = nltk.ChartParser(word)

sent = ['나','는','항상','배','가','고프','았','어요']

for tree in parser.parse(sent):
  print(tree)
```
